```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=20, n_redundant=0, n_classes=2, n_clusters_per_class=1, weights=[0.90], random_state=1, flip_y=0)
```

```python
X1, y1 = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
# summarize class distribution
counter = Counter(y1)
print(counter)
# scatter plot of examples by class label
for label, _ in counter.items():
	row_ix = np.where(y1 == label)[0]
	plt.scatter(X1[row_ix, 0], X1[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
oversample= SMOTE()
X1_transformed, y1_transformed = oversample.fit_resample(X1, y1)
```

```python
counter_new = Counter(y1_transformed)
print(counter_new)
```

```python
counter_new.items()
```

```python
for label, _ in counter_new.items():
	row_ix = np.where(y1_transformed == label)[0]
	plt.scatter(X1_transformed[row_ix, 0], X1_transformed[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```
