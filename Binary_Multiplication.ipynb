```python
def binary_multiplication(num1:str, num2:str)->str:
    """
    Computes the product of two binary numbers. 
    
    Args:
    ----------
    A (str) : A = (A_M-1)2^(M-1) + ... + (A_2)2^2 + (A_1)2^1 + (A_0)2^0  => A = A_M-1A_M-2....A_2A_1A_0  (M-bit binary number) 
    B (str) : B = (B_N-1)2^(N-1) + ... + (B_2)2^2 + (B_1)2^1 + (B_0)2^0  => B = B_N-1B_N-2....B_2B_1B_0  (N-bit binary number)
    
    Returns:
    ----------
    A*B (str) : A*B = A(B_N-1)2^(N-1) + ... + A(B_1)2^1 + A(B_0)2^0 
    """
    
    partial_products = []
    for s2 in num2:
        partial_sums = []
        for s1 in num1:
            partial_sum = int(s1)*int(s2)
            partial_sums.append(partial_sum)
        partial_products.append(partial_sums)
        
    final = []
    for _lists in partial_products:
        result = list(map(lambda x:str(x), _lists))
        final.append(result)
        
    final = final[::-1]
    for i, l in zip(range(1,len(final)), final[1:]):
        l += ['0']*(i)
    final = final[::-1]
    for j in range(1,len(final)):
        final[j] = ['0']*(j) + final[j]
        
    sums = [''.join(element) for element in final]
    
    final_sum = 0
    for i in range(len(sums)):
        final_sum += int(sums[i],2)
    final_sum = bin(final_sum)[2:]
    
    return final_sum
```

```python
a = bin(int(input("Number-1: ")))[2:]
b = bin(int(input("Number-2: ")))[2:]

product = binary_multiplication(a,b)
print(f"The product of {int(a,2)} and {int(b, 2)} is: {int(product, 2)}")
```
